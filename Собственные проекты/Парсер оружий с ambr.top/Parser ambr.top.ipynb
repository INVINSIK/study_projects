```python
def getPageLinks(link, driv):
    """
        Возвращает список ссылок на мемы, полученный с текущей страницы
        
        page_number: int/string
            номер страницы для парсинга
            
    """
    
    driv.get(link)
    time.sleep(2)
    last_height = driv.execute_script("return document.body.scrollHeight")
    while True:
        driv.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)  # Необходимо дать время для загрузки страницы через AJAX.
        new_height = driv.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break  # Если высота страницы не меняется, это означает, что прокрутка достигла низа.
        last_height = new_height
               
    html = driv.page_source

    soup = BeautifulSoup(html, 'html.parser')
    
    # ищем ссылки на мемы и очищаем их от ненужных тэгов
    weapon_links = soup.findAll(lambda tag: tag.name == 'a' and tag.get('class') == ['inline-block'])
    weapon_links = ['https://gi.yatta.moe' + link.attrs['href'] for link in weapon_links]
    
    return weapon_links
```

```python
def getStats(sup, find, stats):
    """
        Возвращает очищенное тип оружия просмотров/базовую атаку/...
        
        soup: объект bs4.BeautifulSoup 
            представление текущей страницы
            
        stats: string
            weapon_type, base_attack, main_stat, main_stat_quantity, description, lore
            
    """
    try:
        obj = sup.find_all('div', attrs={'class':stats})
        obj = obj[find - 1]
        obj = obj.text.strip()
    except:
        obj=None
    
    return obj
```

```python
weapon_type = getStats(soup, 1, stats='font-whitney text-lg')
base_attack = getStats(soup, 1, stats='text-xl lg:text-2xl xl:text-3xl')
main_stat = getStats(soup, 2, stats='text-white text-opacity-70 md:text-lg lg:text-lg xl:text-xl')
main_stat_quantity = getStats(soup, 2, stats='text-xl lg:text-2xl xl:text-3xl')
passive_r1 = getStats(soup, 1, stats ='font-bold text-amberBlack select-text')
passive_r5 = getStats(soup2, 1, stats ='font-bold text-amberBlack select-text')
description = getStats(soup, 1, stats='select-text font-whitney text-amberGray select-text')
lore = getStats(soup, 1, stats='flex select-text flex-col gap-4 whitespace-pre-wrap text-white lg:text-lg')

print(f'weapon_type: {weapon_type}\nbase_attack: {base_attack}\nmain_stat: {main_stat}\nmain_stat_quantity: {main_stat_quantity}\npassive_r1: {passive_r1}\npassive_r5: {passive_r5}\ndescription: {description}\nlore: {lore}')
```

```python
def getMaterials(soup):
    
    """
        Возвращает количество материалов, необходимых для полного возвышения оружия
        
        soup: объект bs4.BeautifulSoup 
            представление текущей страницы
            
    """
    
    materials = soup.find('div', attrs={'class':'flex flex-row flex-wrap justify-center gap-4 xl:justify-start'})
    
    mats = materials.find_all('div', class_= 'absolute z-10 w-full rounded-t-lg bg-amberDark-900 bg-opacity-70 px-1 text-center text-xs opacity-100 md:text-sm lg:text-sm xl:text-sm')
    
    titles2 = materials.find_all('img')
    
    h = ''
    
    for i in range(len(mats)):
        h = h + titles2[i]['title'] + ' - ' + mats[i].text.strip() + ', '
    
    h = h[:-2]
    
    return h
```

```python
def getWeaponData(weapon_link):
    
    """
        Запрашивает данные по странице, возвращает обработанный словарь с данными
        
        meme_page: string
            ссылка на страницу с оружием
    
    """
    driv.get(weapon_link)
    time.sleep(2)
    last_height = driv.execute_script("return document.body.scrollHeight")
    while True:
        driv.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)  # Необходимо дать время для загрузки страницы через AJAX.
        new_height = driv.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break  # Если высота страницы не меняется, это означает, что прокрутка достигла низа.
        last_height = new_height

    html = driv.page_source

    soup = BeautifulSoup(html, 'html.parser')

    time.sleep(1)

    # Предположим, что инициализацию веб-драйвера вы уже осуществили
    try:
        select = Select(driv.find_element(By.XPATH, '//*[@id="app"]/main/div/div[2]/div[1]/div/div[1]/div/div/div[2]/div/div[2]/div[1]/select')) # Это наш выпадающий список, скажем, в меню ресторана
        select.select_by_value("4")  # Выбираем элемент по его значению, представьте, что это особенное блюдо в меню.
        # И тут для нас открываются другие возможности:
        # select.select_by_index(42) # Если мы точно знаем индекс нужного нам пункта.
        # select.select_by_visible_text("Text") # Если мы хотим выбрать элемент по его видимому тексту.
        html2 = driv.page_source

        soup2 = BeautifulSoup(html2, 'html.parser')
    except:
        soup2 = []

    
    # Начало доставания данных
    name = soup.find('h1', attrs={'class':'my-2 font-whitney text-3xl lg:mx-0 xl:mx-0'}).text.strip()
    
    weapon_type = getStats(soup, 1, stats='font-whitney text-lg')
    base_attack = getStats(soup, 1, stats='text-xl lg:text-2xl xl:text-3xl')
    main_stat = getStats(soup, 2, stats='text-white text-opacity-70 md:text-lg lg:text-lg xl:text-xl')
    main_stat_quantity = getStats(soup, 2, stats='text-xl lg:text-2xl xl:text-3xl')
    passive_r1 = getStats(soup, 1, stats ='font-bold text-amberBlack select-text')
    passive_r5 = getStats(soup2, 1, stats ='font-bold text-amberBlack select-text')
    description = getStats(soup, 1, stats='select-text font-whitney text-amberGray select-text')
    lore = getStats(soup, 1, stats='flex select-text flex-col gap-4 whitespace-pre-wrap text-white lg:text-lg')
    
    materials = getMaterials(soup)
    
    data_row = {"Название":name, "Тип оружия":weapon_type, 
                "Базовая атака":base_attack, "Главная характеристика":main_stat, 
                "Колиество стата главной характеристики":main_stat_quantity,
                "Пассивная способность на первом пробуждении":passive_r1, 
                "Пассивная способность на пятом пробуждении":passive_r5, 
                "Описание":description, "История оружия":lore, "Необходимые для прокачки материалы":materials}
    
    return data_row
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from tqdm import tqdm_notebook
from bs4 import BeautifulSoup

final_df = pd.DataFrame(columns = ["Название", "Тип оружия", 
                "Базовая атака", "Главная характеристика", 
                "Колиество стата главной характеристики",
                "Пассивная способность на первом пробуждении", 
                "Пассивная способность на пятом пробуждении", 
                "Описание", "История оружия", "Необходимые для прокачки материалы"])

link = 'https://gi.yatta.moe/ru/archive/weapon'

driv = webdriver.Chrome()
driv.set_window_position(0, 0)
driv.set_window_size(1920, 1080)
driv.get(link)
WebDriverWait(driv, 15).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[2]/div/div[2]/div[1]/div[2]/div/div')))
 
weapon_links = getPageLinks(link, driv)  
    
for weapon_link in tqdm_notebook(weapon_links, desc='Weapons', leave=False):
        
        # иногда с первого раза страничка не прогружается
        for i in range(3):
            try:
                # пытаемся собрать данные
                data_row = getWeaponData(weapon_link)
                final_df = pd.concat([final_df, pd.DataFrame([data_row])], ignore_index=True)
                print(data_row['Название'])
                # если всё получилось - выходим из внутреннего цикла
                break
            except:
                # Иначе, пробуем еще несколько раз, пока не закончатся попытки
                print('AHTUNG! parsing once again:', weapon_link)
                continue

driv.quit()
```
